### Single page scraping

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
url="https://www.flipkart.com/search?q=televisions&sid=ckf%2Cczl&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_psas-pos=1&as-type=RECENT&suggestionId=televisions%7CTelevisions&requestId=4509ecfa-913b-485b-974c-113dd214c08b&as-searchtext=tele"
page=requests.get(url)
soup = BeautifulSoup(page.text)
# prices
p=soup.find_all('div',class_='_30jeq3 _1_WHN1')
Prices=[]
for i in p:
    Prices.append(i.text)

# models
m=soup.find_all('div',class_='_4rR01T')
Model=[]
for i in soup.find_all('div',class_='_4rR01T'):
    Model.append(i.text.split("(")[0].split(" ")[0:1])
    
# Inch
Inch=[]
for i in m:
    a = i.text
    b = re.findall(r"(\d+\sinch)",a)
    if b:
        Inch.append(b)
        
# Operating
op=soup.find_all('li',class_="rgWa7D")
Operating = []
for i in op:
    a = i.text
    b = re.findall(r"\w+\s\w+:\s(\w+)",a)
    if b:
        Operating.append(b)

# UltraHD
UltraHD= []
for i in op:
    a = i.text
    b = re.findall(r"(\d+\sx\s\d+)\sPixels",a)
    if b:
        UltraHD.append(b)

# Speaker
Speaker= []
for i in op:
    a = i.text
    b = re.findall(r"(\d+\sW)\sSpeaker",a)
    if b:
        Speaker.append(b)

# Percentage
per=soup.find_all('div',class_="_3Ay6Sb")
Percentage=[]
for i in per:
    Percentage.append(i.text)

for i in op:
    Speaker.append("nan")

for i in op:
    Operating.append("nan")

dictionary={'Prices':Prices,
           'Model':Model,
           'Inch':Inch,
           'Operating':Operating[0:24],
           'UltraHD':UltraHD,
           'Speaker':Speaker[0:24],
           'Percentage':Percentage[0:24]}

df = pd.DataFrame(dictionary)
df
```

```python
url="https://www.flipkart.com/search?q=televisions&sid=ckf%2Cczl&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_4_4_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_4_4_na_na_ps&as-pos=4&as-type=HISTORY&suggestionId=televisions%7CTelevisions&requestId=4509ecfa-913b-485b-974c-113dd214c08b&page="

Prices=[]
Model=[]
inch=[]
Operating = []
UltraHD=[]
Speaker=[]
percentage=[]

for i in range(1,18):
    urls = url+str(i)
    
    print(urls)
    page = requests.get(urls)
    soup = BeautifulSoup(page.text)

# models
    
    m=soup.find_all('div',class_='_4rR01T')
    for i in soup.find_all('div',class_='_4rR01T'):
        Model.append(i.text.split("(")[0].split(" ")[0:1])
    

# prices
    
    p=soup.find_all('div',class_='_30jeq3 _1_WHN1')
    for i in p:
        Prices.append(i.text)

# Inch
    y=soup.find_all("div",class_="_4rR01T")
    for i in y:
        a = i.text
        b = re.findall(r"(\d+\sinch)",a)
        
        if b:
            inch.append(b)
        
# Operating
    
    op=soup.find_all('li',class_="rgWa7D")
    
    for i in op:
        a = i.text
        b = re.findall(r"\w+\s\w+:\s(\w+)",a)
        if b:
            Operating.append(b)
# UltraHD

    for i in op:
        a = i.text
        b = re.findall(r"(\d+\sx\s\d+)\sPixels",a)
        if b:
            UltraHD.append(b)

# Speaker

    for i in op:
        a = i.text
        b = re.findall(r"(\d+\sW)\sSpeaker",a)
        if b:
            Speaker.append(b)

# Percentage
    per=soup.find_all('div',class_="_3Ay6Sb")
    for i in per:
        percentage.append(i.text)
    
    
```
